```python
def filtros(dataframe, **kwargs):
  filtro: pd.core.frame.Dataframe = dataframe
  for k, v in kwargs.items():
    filtro = dataframe[k] == v
  return dataframe[filtro]
```

```python
filtros(df_Winedata, country='US', province = 'California', region_2 = 'Napa', region_1 = 'Napa Valley', density=1.0)
```

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import scipy.stats as st
import plotly.graph_objects as go
import missingno as msno

from sklearn import preprocessing
from sklearn.metrics import f1_score, confusion_matrix,accuracy_score, classification_report
from scipy import stats
```

```python
train = pd.read_csv('WineQT.csv')
```

```python
train.head()
```

```python
train=train.set_index('pH')
```

```python
fig = go.Figure(data=[go.Pie(labels=train['quality'].value_counts().index, values=train['quality'].value_counts(), hole=.3)])
fig.update_layout(legend_title_text='Quality')
fig.show()
```

```python
correlation = train.corr()
print(correlation['quality'].sort_values(ascending = False),'\n')
```

```python
k= 10
cols = correlation.nlargest(k,'quality')['quality'].index
print(cols)
cm = np.corrcoef(train[cols].values.T)
f , ax = plt.subplots(figsize = (14,12))
sns.heatmap(cm, vmax=.8, linewidths=0.01,square=True,annot=True,cmap='viridis',
            linecolor="white",xticklabels = cols.values ,annot_kws = {'size':12},yticklabels = cols.values)
```
